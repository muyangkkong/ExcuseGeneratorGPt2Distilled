```python
import pandas as pd
import ftfy  # fix text for you (pip install ftfy)

# CSV 파일 불러오기
df = pd.read_csv("Casual_Excuse_File.csv")

# ftfy로 문자열 정제
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].apply(lambda x: ftfy.fix_text(str(x)))

# 저장
df.to_csv("Excuses_Final_Fixed_Casual.csv", index=False)
```
