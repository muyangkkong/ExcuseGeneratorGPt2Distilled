```python
from peft import PeftModel, PeftConfig
from transformers import AutoTokenizer, AutoModelForCausalLM

# 모델 경로 (Google Drive에 저장된 LoRA 모델)
model_path = "/content/drive/My Drive/gpt2-lora-w5h1-continued"

# LoRA 설정 로드
peft_config = PeftConfig.from_pretrained(model_path)

# base 모델과 tokenizer 로드
tokenizer = AutoTokenizer.from_pretrained(peft_config.base_model_name_or_path)
model = AutoModelForCausalLM.from_pretrained(peft_config.base_model_name_or_path)

# LoRA weight 로드
model = PeftModel.from_pretrained(model, model_path)

# pad_token 설정 (GPT-2 계열은 pad_token이 없음)
tokenizer.pad_token = tokenizer.eos_token
```

```python
prompt = (
    "[SCENARIO]\n"
"Who: the class representative\n"
"What: submitting the group project budget\n"
"When: yesterday morning\n"
"Where: during our team meeting\n"
"Why: I misunderstood and thought our team leader had already submitted it.\n"
"How: Could I have an extension to submit it myself?\n"
"[EXCUSE]\n"

)

# 토크나이즈
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# 생성
output = model.generate(
    **inputs,
    max_length=150,         # 길이 충분히 확보
    do_sample=True,         # 샘플링 모드 (무작위성 포함)
    temperature=0.9,        # 창의성 제어
    top_p=0.95,
    num_return_sequences=1,
    pad_token_id=tokenizer.eos_token_id  # 경고 방지
)

# 디코딩
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
